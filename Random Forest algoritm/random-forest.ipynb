## Random Forest
`Random Forest` algoritmi bu `Machine Learning` algoritmidir. Random Forest algoritmi `Ensemble` algoritmlar turiga kiradi. `Ensemble` bu bir nechta modellar birlashtirlib hosil qilingan modellarga nisbatan ishlatiladi. Random Forest algoritmi nega `Ensemble` desangiz bu `Decision Tree` algoritmlarining birlashmasidir. Yuqoridagi rasmga etibor bergan bolsangiz Random Forest bir nechta `Decision Tree` larga bo'lingan. `Decision Tree` ham Machine Learning algoritmidir. `Decision Tree` algoritmi haqida keyinroq gaplashamiz.**

`Random Forest` algoritmi `1995` chi yilda `Tin Kam Xo` tomonidan `Random Subspace Method (Tasodifiy pastki fazo usuli)` asosida yaratilgan. Algoritm kengaytmasi `Leo Breiman` va `Alde Culter` tomonidan yaratilgan. `2006` yilda "Random Forest" nomi ostida ro'yxatdan otkazilgan. 

`Random Forest` - bu juda ko'p funktsiyalarga ega bo'lgan katta ma'lumotlar to'plamini boshqara oladigan kuchli algoritm. Bu, ayniqsa, yetishmayotgan qiymatlar yoki buzilishlari ma'lumotlar ega bo'lgan ma'lumotlar to'plamlari bilan ishlashda foydalidir, chunki u ularni osonlik bilan boshqarishi mumkin. Random, shuningdek, boshqa Machine Learning algoritmlarida keng tarqalgan muammo bo'lgan `overfitting` ga chidamli.



# Decision Tree
   

`Decision Tree` (Qarorlar daraxti) bu klassifikatsiya va regressiya uchun ishlatiladigan Machine Learning mashhur algoritmlaridan biridir! Bu algoritm har bir hususiyat (feature) bo'yicha ma'lum bir shartlarga asosan `tugunlarga` bo'linadi `tugunlar` esa shartga asosan yana bo'linib ketaveradi toki shart tugaguncha yoki maksimal bo'linishlar soni yoki maksimal chuqurligi berilgada bo'linishlar to'xtaydi va ma'lum bir natijani qaytaradi. Osha natija biz bergan muammoning yechimi hisoblanadi. Decision Tree algoritmi asosan chiziqli bo'lmagan yoki kategoriyali muammolar bilan ishlaganda ancha yaxshi natijalarga erishadi. Haddan tashqari train malumotlariga organish `overfitting` ni oldini olish uchun model aniqligiga sezilarli hissa qoshmaydigan malum bir tugunlar olib tashlanadi. ## Random Forest vs Decision Tree





Random Forest va Decision Tree ikkalasi ham klassifikatsiya va regressiya muammolari uchun ishlatiladigan mashhur Machine Learning algoritmlaridir. Biroq, ikkita algoritm o'rtasida ba'zi bir asosiy farqlar mavjud bo'lib, ularni har xil turdagi muammolar uchun yaxshiroq moslashtirladi.

* Decisoin Tree

Decision Tree oddiy, parametrik bo'lmagan algoritm bo'lib, shartlar va ularning mumkin bo'lgan natijalarini daraxtga o'xshash modelini quradi. Decision Tree klassifikatsiya va raqamli ma'lumotlar bilan yaxshi ishlaydi, lekin katta o'lchamli ya'ni high-dimensional ma'lumotlar yoki murakkab chiziqli bo'lmagan ma'lumotlar uchun samarali bo'lmasligi mumkin. Decision Tree haddan tashqari overfitting ga moyil bo'lib, murakkablashishi mumkin, bu esa yangi ma'lumotlar boyicha yomon ishlashga olib kelishi mumkin.

* Random Forest

Random Forest har bir Decision Tree ni yaratishda ma'lumotlar to'plamidan  tasodifiy tanlab oladi va oqitadi. Bu usul orqali overfitting kamaytiriladi. Random Forest Decision Tree ga qaraganda ko'proq vaqt oqitiladi, lekin yangi ma'lumotlarda yuqori aniqlik va yaxshiroq umumiylashtirish ko'rsatkichiga erisha oladi. Random Forest kategorik , katta o'lchamli ya'ni high-dimensional ma'lumotlar va murakkab chiziqli bo'lmagan ma'lumotlar bilan yaxshi ishlaydi.

## Kerakli modullarni yuklash

```python
# Random Forest qurish uchun DecisionTreeRegressor va DecisionTreeClassifier 
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier 
from collections import Counter 
import numpy as np  # Matritsa va massivlar bilan ishlovchi modul
import pandas as pd # Ma'lumotlar bilan ishlovchi modul
```

## Random Forest Classifier

```python
class CustomRandomForestClassifier:
    '''
    Random Forest algoritmi 0 dam
    '''
    def __init__(self, num_trees=25, min_samples_split=2, max_depth=5):
        """
        Konstruktor
        """
        self.num_trees = num_trees
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        # Decision Tree larni saqlash uchun list
        self.decision_trees = []
        
    @staticmethod
    def _sample(X, y):
        '''
        Ma'lumotlarni randmo tanlovchi funksiya
        '''
        n_rows, n_cols = X.shape
        #
        samples = np.random.choice(a=n_rows, size=n_rows, replace=True)
        return X[samples], y[samples]
        
    def fit(self, X, y):
        '''
        Random Forestni train qilish funksiyasi
        
        :param X: np.array, atributlari(features)
        :param y: np.array, klasslari(target)
        :return: None
        '''

        if len(self.decision_trees) > 0:
            self.decision_trees = []
            
        # Random Forestning Decision Tree (daraxtlarini qurish)
        num_built = 0
        while num_built < self.num_trees:
            try:
                clf = DecisionTreeClassifier(
                    min_samples_split=self.min_samples_split,
                    max_depth=self.max_depth
                )
                # Random ma'lumotlarni olish
                _X, _y = self._sample(X, y)
                # Decision Tree larni oqitish 
                clf.fit(_X, _y)
                # Oqitilgan Decision Tree lani listga saqlash
                self.decision_trees.append(clf)
                num_built += 1
            except Exception as e:
                continue
    
    def predict(self, X):
        '''
        Yangi ma'lumotni bashorat qilish uchun predict fuksiyasi
        
        :param X: np.array, yangi ma'lumot predict uchun
        :return: class
        '''
        # Har bir Decision Tree ni predict qilgan qiymatlarini olyapmiz va listga yig'yapmiz
        y = []
        for tree in self.decision_trees:
            y.append(tree.predict(X))
        
        # Eng ko'p bashorat qilingan qiymatlarni topish uchun litni qayta shakllantiryapmiz
        y = np.swapaxes(a=y, axis1=0, axis2=1)
        
        # Oxirgi predict qiymatni qaytaradi
        predictions = []
        for preds in y:
            counter = Counter(preds)
            predictions.append(counter.most_common(1)[0][0])
        return predictions
```

## Modelni oqitish va testlash uchun ma'lumotlar to'plami

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
iris = datasets.load_iris()
X = iris["data"]
y = iris["target"]
labels = iris["target_names"]
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
custom_rf_classifier = CustomRandomForestClassifier()
```

```python
# Modelni oqitish
custom_rf_classifier.fit(X_train_class, y_train_class)
```

## Custom Random Forest Classifier  test 

```python
from sklearn.metrics import accuracy_score
custom_preds_class = custom_rf_classifier.predict(X_test_class)

print("Accuracy : ", accuracy_score(y_test_class, custom_preds_class))
```

## Random Forest Regressor

```python
class CustomRandomForestRegressor:
    '''
    Random Forest algoritmi 0 dam
    '''
    def __init__(self, num_trees=25, min_samples_split=2, max_depth=5):
        """
        Konstruktor
        """
        self.num_trees = num_trees
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        # Decision Tree larni saqlash uchun list
        self.decision_trees = []
        
    @staticmethod
    def _sample(X, y):
        '''
        Ma'lumotlarni randmo tanlovchi funksiya
        '''
        n_rows, n_cols = X.shape
        #
        samples = np.random.choice(a=n_rows, size=n_rows, replace=True)
        return X[samples], y[samples]
        
    def fit(self, X, y):
        '''
        Random Forestni train qilish funksiyasi
        
        :param X: np.array, atributlari(features)
        :param y: np.array, klasslari(target)
        :return: None
        '''

        if len(self.decision_trees) > 0:
            self.decision_trees = []
            
        # Random Forestning Decision Tree (daraxtlarini qurish)
        num_built = 0
        while num_built < self.num_trees:
            try:
                clf = DecisionTreeRegressor(
                    min_samples_split=self.min_samples_split,
                    max_depth=self.max_depth
                )
                # Random ma'lumotlarni olish
                _X, _y = self._sample(X, y)
                # Decision Tree larni oqitish 
                clf.fit(_X, _y)
                # Oqitilgan Decision Tree lani listga saqlash
                self.decision_trees.append(clf)
                num_built += 1
            except Exception as e:
                continue
    
    def predict(self, X):
        '''
        Yangi ma'lumotni bashorat qilish uchun predict fuksiyasi
        
        :param X: np.array, yangi ma'lumot predict uchun
        :return: class
        '''
        # Har bir Decision Tree ni predict qilgan qiymatlarini olyapmiz va listga yig'yapmiz
        y = []
        
        for x in X:
            y_x = []
            for tree in self.decision_trees:
                y_x.append(tree.predict([x]))
            y.append(np.array(y_x).mean())
        
        return np.array(y)
```

## Modelni oqitish va testlash uchun ma'lumotlar to'plami

```python
custom_rf_regressor = CustomRandomForestRegressor()
```

```python
# Modelni oqitish
custom_rf_regressor.fit(X_train_reg, y_train_reg)
```
