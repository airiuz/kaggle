```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test) = tf.keras.datasets.cifar10.load_data()
```

```python
x_train = x_train.astype(np.float32)/255.0
x_test = x_test.astype(np.float32)/255.0
```

```python
rasm = cv2.imread("dataset/cnn8.png")
rasm = cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,20))
plt.imshow(rasm)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.InputLayer((32,32,3)),
    tf.keras.layers.Conv2D(32,(3,3),padding="same",activation="relu"),  # 32 32 32
    tf.keras.layers.MaxPool2D((2,2)),   # 16 16 32 
    tf.keras.layers.Conv2D(64,(3,3),padding="same",activation="relu"),  # 16 16 64
    tf.keras.layers.MaxPool2D((2,2)), # 8 8 64
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64,activation="relu"),
    tf.keras.layers.Dense(10,activation="softmax")
])
```

```python
rasm = cv2.imread("dataset/cnn10.png")
rasm = cv2.cvtColor(rasm,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,15))
plt.imshow(rasm)
```

```python
model.summary()
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy()
optim = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=optim,loss=loss,metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=10)
```

* Buning natijasida konvolved xarita paydo bo'ladi 
* Konvolved xarita tuzatilgan xususiyat xaritasini yaratish uchun ReLU funktsiyasiga qo'llaniladi 
* Tasvir bir nechta konvolyutsiyalar va xususiyatlarni aniqlash uchun ReLU qatlamlari bilan qayta ishlanadi 
* Tasvirning o'ziga xos qismlarini aniqlash uchun turli filtrlarga ega turli xil birlashtiruvchi qatlamlar qo'llaniladi 
* Birlashtirilgan xususiyat xaritasi tekislanadi va yakuniy natijani olish uchun to'liq bog'langan qatlamga beriladi
