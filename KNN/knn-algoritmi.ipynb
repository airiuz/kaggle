## KNN algoritmi qanday ishlaydi?

#### Yuqorida ko'rganimizdek, KNN algoritmi klassifikatsiya va regressiya masalalari uchun ham qo'llanilishi mumkin. KNN algoritmi har qanday yangi ma'lumotlar nuqtalarining qiymatlarini bashorat qilish uchun " xususiyatlar o'xshashligi " dan foydalanadi. Bu shuni anglatadiki, yangi nuqtaga o'quv majmuasidagi nuqtalarga qanchalik o'xshashligiga qarab qiymat beriladi. Bizning misolimizdan bilamizki, ID11 bo'yi va yoshi ID1 va ID5 ga o'xshash, shuning uchun vazn ham taxminan bir xil bo'ladi.

#### Agar bu klassifikatsiya muammosi bo'lganida, biz oxirgi bashorat sifatida ko'p ovoz berilganini qabul qilgan bo'lardik. Bu holda, biz og'irlikning ikkita qiymatiga egamiz - 72 va 77. Qiymatlarning o'rtacha qiymati yakuniy bashorat sifatida qabul qilinadi(Regressiya uchun).

#### Quyida algoritmning bosqichma-bosqich tushuntirishi keltirilgan:
#### 1. Birinchidan, yangi nuqta va har bir mashg'ulot nuqtasi orasidagi masofa hisoblab chiqiladi.


### 2. Eng yaqin k ma'lumot nuqtasi tanlanadi (masofaga qarab). Ushbu misolda, k ning qiymati 3 bo'lsa, 1, 5 va 6 nuqtalar tanlanadi. 


#### 3. Ushbu ma'lumotlar nuqtalarining o'rtacha qiymati yangi nuqta uchun yakuniy bashoratdir. Bu erda biz ID11 = (77+72+60)/3 = 69,66 kg vaznga egamiz.

### Nuqtalar orasidagi masofa qanday hisoblanadi?

#### Birinchi qadam yangi nuqta va har bir o'quv nuqtasi orasidagi masofani hisoblashdir. Ushbu masofani hisoblashning turli usullari mavjud, ulardan eng ko'p ma'lum bo'lgan usullar - Evklid, Manxetten (uzluksiz) va Hamming masofasi (kategorik uchun).
* #### Evklid masofasi: Evklid masofasi yangi nuqta (x) va mavjud nuqta (y) o'rtasidagi kvadratik farqlar yig'indisining kvadrat ildizi sifatida hisoblanadi.
* #### Manxetten masofasi : Bu haqiqiy vektorlar orasidagi masofa, ularning mutlaq farqlari yig'indisi.Evklid va Manxetten masofa formulalari

* #### Hamming masofasi : U kategorik o'zgaruvchilar uchun ishlatiladi. Agar qiymat (x) va qiymat (y) bir xil bo'lsa, D masofa 0 ga teng bo'ladi. Aks holda D=1.

#### Evklid va Manxetten masofalarining umumlashtirilgan shakli bo'lgan Minkovski masofasi ham mavjud.

#### Yangi kuzatuvning bizning o'quv majmuamizdagi nuqtalardan masofasi o'lchangandan so'ng, keyingi qadam eng yaqin nuqtalarni tanlashdir. Ko'rib chiqiladigan ballar soni k qiymati bilan belgilanadi.

### `Klassifikatsiya uchun - Yuqorida Ko'proq Regressiya masalalari ko'rildi lekin Klassifikatsiya ham huddi shunday jarayon boradi. Klassifikatsiyaga oid datalar bilan huddi shunday jarayon olib boriladi. Ya'ni biz kiritgan nuqtadan har birigacha bo'lgan Evklid masofalari topiladi va shunga asoslanib shu k ta nuqta biz kiritgan nuqtaga ovoz berish yo'li bilan qaror qabul qiladi. Binary classificatsiya uchun True yoki False deb ovoz beradi va oxirida berilgan ovozlardan qay biri ko'pligiga qarab True yoki False qiymat qaytaradi.`

```python
import numpy as np
import pandas as pd
```

### Regressiya

```python
data1 = pd.read_csv("/kaggle/input/knn-datasets/bmi.csv")
data1
```

```python
data1 = data1[["Life expectancy","BMI"]]
data1 = data1.head(20)
data1
```

#### Bizda bu yerda maqsad Life expectancy(O'rtacha umr ko'rish yoshini kiritsak) u bizga KNN dan foydalanib BMI(sog'lomlik koefitsienti) ni qaytarishi kerak bo'ladi

```python
def knn_regressiya(predict):
    k = int((len(data1)**0.5)) # k shunday tanlanishi kerak
    print(f"k = {k}")
    points = abs(data1["Life expectancy"] - predict) # Manxetten formulasiga ko'ra har bir nuqta bn biz kiritgan nuqta orasidagi masofalar topiladi
    k_nearest_neighbors = points.sort_values()[:k] # Eng yaqin k ta qo'shnisi tanlab olinadi
    k_nearest_neighbors_labels = data1.loc[k_nearest_neighbors.index,"BMI"] # Eng yaqin k ta qo'shnisining Labellari olinadi?
    return np.mean(k_nearest_neighbors_labels) # ularning o'rtachasi biz kiritgan son uchun BMI ko'rsatkichi bo'ladi
```

```python
knn_regressiya(predict=60)
```

### Klassifikatsiya

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
```

```python
data2 = pd.read_csv("/kaggle/input/knn-datasets/DTree.csv")
data2 
```

```python
ordinal = OrdinalEncoder() # x larni encodlab beradi
```

```python
datax = data2.drop(["Oyinchilar","Shamol"],axis=1) # faqat x larni o'zini olyapmiz
datax
```

```python
ordinal.fit(datax) # fit qilganimizda datani o'rgandi
```

```python
ordinal.categories_ # o'ziga bitta shablon qivoldi shu asosida classificatsiya qiladi
```

```python
datax = ordinal.transform(datax) # Endi classifikatsiyaga o'tqazdi. 0 dan yozsak ham huddi shunday qilamiz ya'ni ma'lum 
datax                            # bir qoida asosida klassifikatsiyaga o'giramiz
```

```python
def knn_klassifikatsiya(predict):
    k = int((len(data2)**0.5)) # k shunday tanlanishi kerak
    if k%2==0 and k>1:
        k -= 1 # klassifikatsiyada k toq bo'lishi kerak chunki ovozlar bir biriga teng bo'lib qolishi mumkin bu yerda +-1 qilinadi
    print(f"k = {k}")
    points = np.sqrt(np.sum((datax - predict)**2,axis=1)) # Evklid formulasiga ko'ra har bir nuqta bn biz kiritgan nuqta orasidagi masofalar topiladi
    k_nearest_neighbors = pd.DataFrame(points,columns=["yaqinlari"]).sort_values(by="yaqinlari")[:k] # Eng yaqin k ta qo'shnisi tanlab olinadi
    k_nearest_neighbors_labels = data2.loc[k_nearest_neighbors.index,"Shamol"] # Eng yaqin k ta qo'shnisining Labellari olinadi?
    return k_nearest_neighbors_labels.max() # qaysi bir klassga ovoz ko'p bo'lsa o'sha klassni qaytaradi
```
